```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
df = pd.read_csv('estaciones.csv', index_col=None)

Estaciones = nx.from_pandas_edgelist(df,source='Origen',target='Destino',edge_attr='Longitud')
Estaciones.nodes()
```

```python
Estaciones.edges()
```

```python
djk_path=nx.dijkstra_path(Estaciones, source='E7',target='Ladron',weight=True)
djk_path
```

```python
ruta1 = Estaciones.subgraph(['E7', 'E6', 'Sauz', 'Ladron'])
nx.draw(ruta1,with_labels=True)
```
