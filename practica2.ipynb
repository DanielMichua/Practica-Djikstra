```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
#import numpy as np
```

```python
df = pd.read_csv('estaciones.csv', index_col=None)

Estaciones = nx.from_pandas_edgelist(df,source='Origen',target='Destino',edge_attr='Longitud')
```

```python
Estaciones.edges()
```

```python
Estaciones.nodes()
```

```python
Estaciones.order()
```

```python
for x in Estaciones.nodes():
  if Estaciones.degree(x) > 2:
    print(x)
```

```python
rut = nx.dijkstra_path(Estaciones, source='E3',target='Sauz',weight=True)
rut

def plot_shortest_path(rut):

  positions = nx.circular_layout(Estaciones)
  nx.draw(Estaciones, pos = positions, with_labels=True, node_size =500, width = 2, node_color = 'lightblue', edge_color = 'gray')

  short_path = nx.DiGraph()

  for i in range(len(rut)-1):
    short_path.add_edge(rut[i],rut[i+1])

  nx.draw(short_path, pos = positions, with_labels = True, node_size = 500, width = 4, node_color = '#ff5733',edge_color = '#f0b206')


plt.show()

plot_shortest_path(rut)

print(rut)

```
